**Idea Transformerów**

Klasy typu transformers są tworzone aby ułatwić proces np. przetwarzania danych. Zamiast zapisywać dane do zmiennych i przekazywać je dalej, można stworzyć potok (pipeline), aby wyjście było od razu wejściem do następnej operacji.
Własna klasa, która może działać jak transformer musi posiadać następujące metody:

*   __init__(self, arguments)
*   fit(self, X, y=None)
*   transform(self, X)

Obiekty klas, które są transformerami można osadzić w klasie Pipeline, które odpowiedzialna jest za wykonanie potoku. Na każdym obiekcie wywoływana jest metoda fit_transform, który wywołuje w sobie fit a następnie transform.

Z użyciem Transformerów

````python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline


class SelectColumns(BaseEstimator, TransformerMixin):
  def __init__(self, column_names):
    self.column_names=column_names
  def fit(self, X, y=None):
    return self
  def transform(self, X):
    return X[self.column_names]

class RenameColumns(BaseEstimator, TransformerMixin):
  def __init__(self, rename_info):
    self.rename_info=rename_info
  def fit(self, X, y=None):
    return self
  def transform(self, X):
    X_=X.copy()
    return X_.rename(columns=self.rename_info)

class ChooseObservations(BaseEstimator, TransformerMixin):
  def __init__(self, more_than):
    self.more_than=more_than
  def fit(self, X, y=None):
    return self
  def transform(self, X):
    self.mask=X["NumberFamilyMemebrs"]>self.more_than
    return X[self.mask]
````

````python
transformer=Pipeline(
    steps=[("SelectColumns", SelectColumns(selected_columns_name)),
    ("RenameColumns", RenameColumns(rename_cols)),
    ("ChooseObservations", ChooseObservations(5))])

data_after_transformer=transformer.fit_transform(data)
data_after_transformer.head(3)
````

**Outliery w danych - jakie je zidentyfikować?**
-  za pomocą statystyk: mediana, odchylenie standardowe, kwartyli, percentyli
-  za pomocą reguły 1,5 wartości roztępu międzykwartylowego (Q3-Q1) - zwizualizować można za pomocą wykresów typu boxplot.
-  za pomocą reguły 3 sigm (odchyleń) ale tylko dla zmiennych z rozkładem normalnym. Za outliery brane są wartości, które nie nieleżą do przedziału [średnia-3*odchylenia, średnia+3*odchylenia]
-  za pomocą algorytmów uczenia maszynowego: las izolacyjny, Local Outlier Factor
