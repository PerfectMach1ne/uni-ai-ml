## "Zadanie 1." (data manipulation)

```python
import numpy as np
import pandas as pd
import os
from pathlib import Path

current_path = Path(os.path.dirname(os.path.abspath("__file__")))
file_name = "TravelInsurancePrediction.csv"

df=pd.read_csv(os.path.join(current_path,f"{file_name}"))
df.head(7)
```

### 1.)

```python
print("Średnia arytmetyczna wieku klientów:\nμ = " + str(df["Age"].mean()))
print("Zaokrąglona śr. arytmetyczna wieku klientów:\nμ ~= " + str(df["Age"].mean().round()))
```

### 2.)

```python
mask = df["FrequentFlyer"] == "Yes"
print("Mediana przychodu klientów o statusie Frequent Flyer:\nmedian = " + str(df["AnnualIncome"][mask].median()))
```

### 3.)

```python
mask = df["ChronicDiseases"] == 1
print("Ilość klientów z chorobami przewlekłymi:\nC = " + str(df[:][mask].count().unique()[0]))
```

### 4.)

```python
df = df.drop("Unnamed: 0", axis=1)
df.head(3)
```

### 5.)

```python
column_names = ["Age", "AnnualIncome", "FamilyMembers", "ChronicDiseases"]
data_to_save = df[column_names]
data_to_save.to_csv(os.path.join(current_path,"saved_data.csv"), index=False)
```

### 6.)

```python
mask = df["TravelInsurance"] == True
columns = df.columns.values
columns = np.delete(columns, np.where(columns == "TravelInsurance"), axis=0)

data_to_save = df[mask][columns]
data_to_save.to_csv(os.path.join(current_path,"saved_data2.csv"), index=False)
```

### 7.)

```python
mean = df["AnnualIncome"].mean()
mask = df["AnnualIncome"]>mean
print(f"Klienci, których przychód jest większy niż średnia ({mean}):")
df[mask]
```

### 8.)

```python
new_df = df.copy()
new_df.insert(len(new_df.columns), "AgeGroup", np.nan, allow_duplicates=True)
new_df['AgeGroup'] = new_df['AgeGroup'].astype("str")
new_df["AgeGroup"][new_df["Age"]<30] = "less 30"
new_df["AgeGroup"][(new_df["Age"]>=30) & (new_df["Age"]<40)] = "30-39"
new_df["AgeGroup"][(new_df["Age"]>=40) & (new_df["Age"]<49)] = "40-49"
new_df["AgeGroup"][new_df["Age"]>50] = "50+"
new_df
```

```python
new_df["AgeGroup"]=new_df["AgeGroup"].astype("category")
print(new_df.dtypes)
```
